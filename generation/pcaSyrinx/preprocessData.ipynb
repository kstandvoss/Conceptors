```python
%matplotlib inline

import os
import wave #wav-files whoop!
import matplotlib.pyplot as plt
import numpy as np

#may be changed to point to the syll folder on your system
path = '../../birddb/syll/'

counter = 0
maxNumFiles = 10

for root, dirs, files in os.walk(path):
    for name in files:
        if name.endswith(".wav") and counter < maxNumFiles:
            counter = counter + 1;
            print(os.path.join(root,name))
            #read in file and determine length
            w = wave.open(os.path.join(root,name), 'r')
            wavParams = w.getparams()
            print(wavParams.nframes) 
            
            #Extract Raw Audio from Wav File
            signal = w.readframes(-1)
            signal = np.fromstring(signal, 'Int16')
            fs = w.getframerate()

            time = np.linspace(0, len(signal)/fs, num=len(signal))
            
            print(time)
            
            plt.figure(counter)
            plt.title('Signal Wave for '+ name)
            plt.plot(time,signal)
            
```
