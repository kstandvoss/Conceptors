```python
import textgrid as t
import wave
import numpy as np
import os
import re
```

```python
os.chdir('/home/felix/Schreibtisch/BirdDB-Test')

# open csv to get filnames of textgrid files
filenames = np.recfromcsv('cavi_all_textgridpath.csv', delimiter=',')

# change to directory of BirdDB 
os.chdir('/media/felix/TOSHIBA EXT/BirdDB/all')

totalcount = 0

# for all files that are in the list
for myfile in enumerate(filenames):
    # dismiss the path, just get the filename
    textgridfile = re.split('/', myfile[1][0])
    # get rid of .TextGrid extension also
    filename = re.split('\.', textgridfile[-1:][0])
    filename = filename[0] 

    # open the file
    if os.path.isfile(filename + '.WAV') and os.access(filename + '.WAV', os.R_OK):
        w = wave.open(filename + '.WAV')
        print 'Opened ' + filename + '.WAV'
        
        params = w.getparams()
        #sample freq 
        freq = params[2]
        
        # open the corresponding TextGrid file
        if os.path.isfile(filename + '.TextGrid') and os.access(filename  + '.TextGrid', os.R_OK):
            f = open(filename + '.TextGrid','r')
            #print 'Opened ' + filename + '.TextGrid'
        
            fid = t.TextGrid(f.read())
            
            c = 0
            for i, tier in enumerate(fid):             
                trans = tier.make_simple_transcript()
                # search for something in the list
                matches = [x for x in trans if x[2] != "" and len(x[2]) == 2]   
                
                print 'Found ' + str(len(matches)) + ' matching syllables'
    

                # get soundwave of the items in the search result
                for item in matches:
                    #space to store the data of one syllable
                    sylls = []
                    
                    start_pos = int(float(item[0]) * freq)
                    stop_pos = int(float(item[1]) * freq)
                    syll_name = item[2]

                    w.setpos(start_pos)
                    
                    # get syllable data
                    sample = w.readframes(stop_pos-start_pos)
                    waveform = np.fromstring(sample, 'Int16')
                    # store syllable data 
                    sylls.append(waveform.tolist())
                    
                    # get signal to  the right  format
                    C = [item for sublist in sylls for item in sublist]
                    signal = "".join((wave.struct.pack('h', item) for item in C))

                    os.chdir('/media/felix/TOSHIBA EXT/BirdDB/syll/')
                    directory = os.path.join(os.getcwd(), syll_name)
                    # choose the right folder
                    if not(os.path.isdir(directory)):
                        os.makedirs(directory)
                    os.chdir(directory)

                           
                    # prepate output file
                    number = str(c)
                    output = wave.open(syll_name + '_' + str(c) + '_' + filename + '.wav', 'w')
                    output.setparams((1, 2, 44100, 0, 'NONE', 'not compressed'))
                    c = c + 1

                    # write to file
                    output.writeframes(signal)
                    # print 'wrote ' + syll_name + '_' + str(c) + '.wav'
                    totalcount = totalcount + 1
                    output.close()
        
                    # change to directory of BirdDB 
                    os.chdir('/media/felix/TOSHIBA EXT/BirdDB/all')
            f.close()
        else:
            print filename + '.TextGrid' + "Either file is missing or is not readable"
         
        w.close()
    else:
        print filename  + '.WAV' + "Either file is missing or is not readable"
    
print 'Wrote ' + str(totalcount) + " Syllable files"
```
